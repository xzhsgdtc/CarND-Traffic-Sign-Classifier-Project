````python

def print_info(i):
    with tf.Session() as sess:
        output = sess.run(i)
        print('value is ',i,'\ntype is ',type(output))
````

````python
A = tf.constant([12,12,123])
B = tf.constant([[1,2,3],[3,3,4],[3,2,2]])
print_info(A)
print_info(B)
````

````python
import numpy as np
def soft_max(l):
    e_sum = sum([np.exp(i) for i in l])
    return [np.exp(i)/e_sum for i in l]
l = [1.0,2,3]
l1 = [2,4,6]
print(soft_max(l),'\n',soft_max(l1))
tf.nn.softmax(l)
````
